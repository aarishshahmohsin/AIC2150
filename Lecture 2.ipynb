```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print(f"Feature names: {iris_dataset['feature_names']}")
```

```python
print(f"The size of the data is: {iris_dataset['data'].shape}")
```

```python
# 0 setosa, 1 Versicolor, 2 Virgiica
print(f"The target values are: \n{iris_dataset['target']}")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], test_size = 0.2, random_state= 0)
```

```python
print(f"X_train shape: {X_train.shape}")
```

```python
print(f"X_test shape: {X_test.shape}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns= iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, c=y_train ,figsize=(15,15) ,hist_kwds={'bins':20}, s=60,alpha=0.8)
```

```python
#importing knn classifier 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
# .fit method for training
knn.fit(X_train, y_train)
```
